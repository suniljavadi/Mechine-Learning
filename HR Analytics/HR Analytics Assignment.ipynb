```python
#data visualization, analyzing different features

ftr=["dept","salary","numberOfProjects","timeSpent.company","workAccident","promotionInLast5years"]
```

```python
ftr
```

```python
fig=plt.subplots(figsize=(20,25)) #to plot multiple plots

for p,q in enumerate(ftr):
    plt.subplot(3,2,p+1)
    plt.subplots_adjust(hspace=0.5)
    sns.countplot(x=q,data=data,hue="left")
    plt.xticks(rotation=90)
```
